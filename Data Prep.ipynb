## Create Homogenous Segments

```python
multilane = pd.read_csv('Clean Data/multilane.csv')
multilane.head()
```

```python
#horizontal curvature 
multilane['DegHCurve_LWA'] = ((multilane['CU_A_LENGTH']*1.7)+(multilane['CU_B_LENGTH']*4.45)+
                          (multilane['CU_C_LENGTH']*6.95)+(multilane['CU_D_LENGTH']*11.2)+
                          (multilane['CU_E_LENGTH']*20.95)+(multilane['CU_F_LENGTH']*30))/(multilane['SECTION_LENGTH'])
```

```python
#vertical grade
multilane['DegGrade_LWA'] = ((multilane['GR_A_LENGTH']*0.2)+(multilane['GR_B_LENGTH']*1.45)+
                          (multilane['GR_C_LENGTH']*3.45)+(multilane['GR_D_LENGTH']*5.45)+
                          (multilane['GR_E_LENGTH']*7.45)+(multilane['GR_F_LENGTH']*9.45))/(multilane['SECTION_LENGTH'])
```

```python
multilane['TRUCK_PERC'] = multilane['AVG_SINGLE_UNIT'] + multilane['AVG_COMBINATION']
```

```python
ln = multilane.sort_values(by=['ROUTE_ID', 'BEGIN_POINT', 'END_POINT'], ascending = True).reset_index(drop=True)
ln['LNID'] = ln.index
```

```python
print(len(ln))
ln.tail()
```

```python
revid = 1
i = 0
while i < len(ln)-1:
    ln.loc[i, 'LNSegID'] = revid
    if ln.loc[i, 'ROUTE_ID'] == ln.loc[i+1, 'ROUTE_ID']:
        if (ln.loc[i, 'THROUGH_LANES'] != ln.loc[i+1, 'THROUGH_LANES']) |\
           (ln.loc[i, 'FS_SYSTEM'] != ln.loc[i+1, 'FS_SYSTEM']) |\
           (ln.loc[i, 'RURAL_URBAN'] != ln.loc[i+1, 'RURAL_URBAN']) |\
           (ln.loc[i, 'DegHCurve_LWA'] != ln.loc[i+1, 'DegHCurve_LWA']) |\
           (ln.loc[i, 'DegGrade_LWA'] != ln.loc[i+1, 'DegGrade_LWA']) |\
           (ln.loc[i, 'AADT'] != ln.loc[i+1, 'AADT']) |\
           (ln.loc[i, 'LANE_WIDTH'] != ln.loc[i+1, 'LANE_WIDTH']) |\
           (ln.loc[i, 'SPEED_LIMIT_LWA'] != ln.loc[i+1, 'SPEED_LIMIT_LWA']) |\
           (ln.loc[i, 'SHLD_WIDTH_L'] != ln.loc[i+1, 'SHLD_WIDTH_L']) |\
           (ln.loc[i, 'SHLD_WIDTH_R'] != ln.loc[i+1, 'SHLD_WIDTH_R']) |\
           (ln.loc[i, 'TYPE_TERRAIN'] != ln.loc[i+1, 'TYPE_TERRAIN']) |\
           (ln.loc[i, 'MEDIAN_TYPE'] != ln.loc[i+1, 'MEDIAN_TYPE']) |\
           (ln.loc[i, 'MEDIAN_WIDTH'] != ln.loc[i+1, 'MEDIAN_WIDTH']) |\
           (ln.loc[i, 'ACCESS_CONTROL'] != ln.loc[i+1, 'ACCESS_CONTROL']) |\
           (ln.loc[i, 'PEAK_CAPACITY_HCM6'] != ln.loc[i+1, 'PEAK_CAPACITY_HCM6']) |\
           (ln.loc[i, 'TRUCK_PERC'] != ln.loc[i+1, 'TRUCK_PERC']):
            revid += 1
    else:
        revid += 1
    i += 1
ln.loc[i, 'LNSegID'] = revid
```

```python
ln.tail()
```

```python
first = lambda x: x.iloc[0]
lwa = lambda x: np.average(x, weights=ln.loc[x.index, "SECTION_LENGTH"])

ln_grouped = ln.groupby('LNSegID', as_index=False).agg({'ROUTE_ID':first,
                                          'BEGIN_POINT': 'min',
                                          'END_POINT': 'max',
                                          'LNID' : 'nunique',            
                                          'RURAL_URBAN':first,
                                          'FS_SYSTEM': lwa,    
                                          
                                          
                                          'INTERCHANGES': 'sum', 
                                          'AT_GRADE_OTHER': 'sum',
                                          'AT_GRADE_SIGNALS': 'sum',
                                          'AT_GRADE_SIGNS': 'sum', 
                                          'ACCESS_CONTROL':lwa,
                                           
                                          'AVG_COMBINATION': lwa,               
                                          'AVG_SINGLE_UNIT': lwa,       
                                          'PERC_COMBINATION': lwa,              
                                          'PERC_SINGLE_UNIT': lwa,  
                                                        
                                          'MEDIAN_TYPE': lwa,
                                          'MEDIAN_WIDTH': lwa,
                                          'SHLD_WIDTH_L':lwa,
                                          'SHLD_WIDTH_R':lwa,
                                                        
                                          
                                                        
                                          'DegHCurve_LWA':lwa,
                                          'DegGrade_LWA':lwa,
                                                        
                                          'AADT': lwa,
                                          'TRUCK_PERC': lwa,
                                          'TYPE_TERRAIN':lwa,
                                          'THROUGH_LANES': lwa,
                                          'LANE_WIDTH': lwa,
                                          'SPEED_LIMIT_LWA':lwa,
                    
                                          'TotalCrashes':'sum',
                                          'TotalKilled':'sum',
                                          'TotalInjury' :'sum',    
                                          'Total_A' :'sum',  
                                          'Total_B' :'sum',                 
                                          'Total_C' :'sum',
                                          'TotalNonInjury':'sum',
                                                        
                                          'PEAK_CAPACITY_HCM6':lwa,
                                          'ESTIMATED_VSF_HCM6':lwa,
                                          })
```

```python
ln_grouped["SECTION_LENGTH"] =  abs(ln_grouped['END_POINT'] - ln_grouped['BEGIN_POINT'])
ln_grouped.head()
```
